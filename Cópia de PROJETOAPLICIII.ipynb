```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics.pairwise import cosine_similarity

# Carregar a base de dados
df = pd.read_csv("cursos_gratuitos.csv")

# Exibir informações iniciais sobre os dados
print("Resumo dos Dados:")
print(df.info())
print("\nVisualização das primeiras linhas:")
print(df.head())

# Transformar colunas categóricas em numéricas
label_encoder = LabelEncoder()
df['area_encoded'] = label_encoder.fit_transform(df['area'])
df['plataforma_encoded'] = label_encoder.fit_transform(df['plataforma'])

# Criar matriz de características
features = df[['area_encoded', 'plataforma_encoded', 'avaliacao']]

# Calcular similaridade do cosseno
similaridade = cosine_similarity(features)

# Criar função de recomendação
def recomendar_cursos(curso_index, num_recomendacoes=5):
    similaridades = list(enumerate(similaridade[curso_index]))
    similaridades = sorted(similaridades, key=lambda x: x[1], reverse=True)[1:num_recomendacoes+1]
    indices_recomendados = [i[0] for i in similaridades]
    return df.iloc[indices_recomendados]

# Testar recomendação
curso_teste = 0  # Índice do curso para testar a recomendação
print("Cursos recomendados para:")
print(df.iloc[curso_teste])
print("\nRecomendações:")
print(recomendar_cursos(curso_teste))

# Visualização das avaliações
df['avaliacao'].hist(bins=10)
plt.xlabel('Avaliação')
plt.ylabel('Quantidade de Cursos')
plt.title('Distribuição das Avaliações')
plt.show()

print("Modelo de recomendação inicial testado com sucesso!")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar a base de dados
df = pd.read_csv("cursos_gratuitos.csv")

# Calcular a média das avaliações por plataforma
df_plataformas = df.groupby('plataforma')['avaliacao'].mean().reset_index()

# Selecionar as 10 melhores plataformas
df_top_plataformas = df_plataformas.nlargest(10, 'avaliacao')

# Criar gráfico de ranking das plataformas
plt.figure(figsize=(10, 6))
sns.barplot(y=df_top_plataformas['plataforma'], x=df_top_plataformas['avaliacao'], palette='viridis')
plt.xlabel("Avaliação Média")
plt.ylabel("Plataforma")
plt.title("Top 10 Plataformas de Cursos Gratuitos Melhor Avaliadas")
plt.xlim(4, 5)  # Ajustando a escala para visualizar melhor
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()

print("Ranking visual das 10 melhores plataformas gerado com sucesso!")
```

```python
# ==========================
# ETAPA 1: Importações
# ==========================
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics.pairwise import cosine_similarity
import seaborn as sns
import matplotlib.pyplot as plt

# ==========================
# ETAPA 2: Carregamento e Pré-processamento
# ==========================
# Carrega o CSV
df = pd.read_csv("cursos_gratuitos.csv")

# Remove colunas codificadas anteriormente, se existirem
df = df.drop(columns=['area_encoded', 'plataforma_encoded'], errors='ignore')

# One-Hot Encoding para 'area' e 'plataforma'
df_encoded = pd.get_dummies(df, columns=['area', 'plataforma'], drop_first=True)

# Normalização da coluna 'avaliacao'
scaler = MinMaxScaler()
df_encoded['avaliacao'] = scaler.fit_transform(df_encoded[['avaliacao']])

# ==========================
# ETAPA 3: Construção do Modelo
# ==========================
# Seleção das features para a matriz de similaridade
features = df_encoded.drop(columns=['nome_do_curso', 'link'])

# Cálculo da similaridade do cosseno
similaridade = cosine_similarity(features)

# ==========================
# ETAPA 4: Função de Recomendação
# ==========================
def recomendar_cursos(curso_index, num_recomendacoes=5):
    similaridades = list(enumerate(similaridade[curso_index]))
    similaridades = sorted(similaridades, key=lambda x: x[1], reverse=True)[1:num_recomendacoes+1]
    indices_recomendados = [i[0] for i in similaridades]
    return df.iloc[indices_recomendados]

# ==========================
# ETAPA 5: Teste da Recomendação
# ==========================
curso_teste = 0  # Índice do curso a ser testado
print("Curso base:")
print(df.iloc[curso_teste])
print("\nCursos recomendados:")
print(recomendar_cursos(curso_teste))

# ==========================
# ETAPA 6: Visualização da Distribuição das Avaliações
# ==========================
df['avaliacao'].hist(bins=10)
plt.xlabel('Avaliação')
plt.ylabel('Quantidade de Cursos')
plt.title('Distribuição das Avaliações dos Cursos')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
